```python
import pandas as pd
```

```python
#New file contains Salary Estimate without PerHour and -1
df = pd.read_csv('my_dataframe.csv')
df
```

```python
def splitter(a):
  a = a.split()
  return(a)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(splitter)
```

```python
def min_finder (a):
  a = a[0]
  return(a)
def max_finder(a):
  a = a[2]
  return(a)
```

```python
df['Min_salary'] = df['Salary Estimate'].apply(min_finder)
df['Max_salary'] = df['Salary Estimate'].apply(max_finder)
```

```python
df['Min_salary'] = df['Min_salary'].astype(int)
df['Max_salary'] = df['Max_salary'].astype(int)
```

```python
def state(a):
  a= a.split()
  a = a[-1]
  return(a)
```

```python
 df['State'] = df['Location'].apply(state)
```

```python
df
```

```python
#Remove Kingdom from state Column
df = df[df['State'] != 'Kingdom']
```

```python
#Remove -1 from Rating Column
df = df[df['Rating']!= -1]
```
